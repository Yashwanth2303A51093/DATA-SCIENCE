```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (file is already uploaded in Colab's file explorer)
df = pd.read_csv("daily_temperature.csv")

# Plot line chart
plt.figure(figsize=(10, 6))
plt.plot(df['Day'], df['Temperature'], marker='o', linestyle='-', color='b', label='Temperature')

# Mark hottest and coldest days
hottest_day = df.loc[df['Temperature'].idxmax()]
coldest_day = df.loc[df['Temperature'].idxmin()]

plt.scatter(hottest_day['Day'], hottest_day['Temperature'], color='red', s=100, label='Hottest Day')
plt.scatter(coldest_day['Day'], coldest_day['Temperature'], color='cyan', s=100, label='Coldest Day')

# Annotate points
plt.text(hottest_day['Day'], hottest_day['Temperature']+0.5,
         f"Hottest: {hottest_day['Temperature']}°C", ha='center', color='red')
plt.text(coldest_day['Day'], coldest_day['Temperature']-1,
         f"Coldest: {coldest_day['Temperature']}°C", ha='center', color='blue')

# Labels and styling
plt.xlabel("Day")
plt.ylabel("Temperature (°C)")
plt.title("Daily Temperature Changes Over a Month")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Print hottest and coldest days
print(f"Hottest Day: Day {hottest_day['Day']} with {hottest_day['Temperature']}°C")
print(f"Coldest Day: Day {coldest_day['Day']} with {coldest_day['Temperature']}°C")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load dataset
df = pd.read_csv("sales_data (1).csv")

# Step 2: Plot line chart
plt.figure(figsize=(8,5))
plt.plot(df["Quarter"], df["Sales"], marker='o', linestyle='-', color='b', label="Sales")

# Highlight peak sales
peak_idx = df["Sales"].idxmax()
plt.scatter(df["Quarter"][peak_idx], df["Sales"][peak_idx], color='red', s=100, label="Peak Sales")

# Labels and title
plt.xlabel("Quarter")
plt.ylabel("Sales (in USD)")
plt.title("Quarterly Sales Performance")
plt.grid(True)
plt.legend()

# Show chart
plt.show()

# Step 3: Interpretation
peak_quarter = df.loc[peak_idx, "Quarter"]
peak_sales = df.loc[peak_idx, "Sales"]

print(f"📈 Peak Sales occurred in {peak_quarter} with sales of ${peak_sales}.")
print("💡 Interpretation: The company experiences the highest sales in the last quarter (Q4),")
print("likely due to festive season demand and year-end promotions.")
```
